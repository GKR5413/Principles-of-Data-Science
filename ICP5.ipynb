```python
num_columns = len(df.columns)
print(f"Number of columns: {num_columns}")
```

```python
mean_house_price = df['price'].mean()
print(f"Mean of house prices: {mean_house_price}")
```

```python
median_house_price = df['price'].median()
print(f"Median of house prices: {median_house_price}")
```

```python
property_types = df['type'].unique()
print(f"Different types of property: {property_types}")
```

```python
condo_df = df[df['type'] == 'Condo']
max_condo_price = condo_df['price'].max()
print(f"Maximum price of a condo: {max_condo_price}")
```

```python
mean_prices_by_type = df.groupby('type')['price'].mean()
print(mean_prices_by_type)
```

```python
property_type_counts = df['type'].value_counts()
print(property_type_counts)
```
